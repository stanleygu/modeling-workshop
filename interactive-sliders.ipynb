```python
def plotResults(result, show=True, legend=False, lineWidth=2.5):
    import matplotlib.pyplot as p

    if result.dtype.names is None:
        # treat as a regular array
        p.plot(result[:,0], result[:,1:])

    else:
        if len(result.dtype.names) < 1:
            raise Exception('no columns to plot')

        time = result.dtype.names[0]

        for name in result.dtype.names[1:]:
            p.plot(result[time], result[name], linewidth=5, label='$' + name + '$')
    
    if legend:
        p.legend(loc = legend)
        
    if show:
        p.show()

def makeParameterSliders(r, plotFn=plotResults, paramMap=None):
    import matplotlib.pyplot as plt
    %matplotlib inline
    %config InlineBackend.figure_format='retina'
    from IPython.html.widgets import interact, interactive, fixed
    from IPython.html import widgets
    from IPython.display import clear_output, display, HTML
    import sys



    def runSim(start=0, stop=100, steps=100, **paramMap):
        r.reset()
        for k,v in paramMap.items():
            try:
                key = k.encode('ascii', 'ignore')
                r.model[key] = v
            except:
                # error in setting model variable
                e = sys.exc_info()
                print e

        try: 
            s = r.simulate(start, stop, steps)
            plotFn(s)
        except:
            # error in simulation
            e = sys.exc_info()
            print e

    if paramMap is None: 
        paramIds = r.model.getGlobalParameterIds()
        paramValues = r.model.getGlobalParameterValues()
        paramMap = {}
        for i, id in enumerate(paramIds):
            val = paramValues[i]
            try:
                r.model[id] = val
                paramMap[id] = widgets.FloatSliderWidget(
                            min=0, max=val*2, step=val/10, value=val)
            except:
                e = sys.exc_info()
                print e

    i = interact(runSim,
             start=widgets.FloatTextWidget(min=0, value=0),
             stop=widgets.FloatTextWidget(min=0, value=100),
             steps=widgets.IntTextWidget(min=0, value=100),
             **paramMap
             )
```
