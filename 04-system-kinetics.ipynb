````python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import tellurium as te
````

# System Kinetics

How systems behave

## Thermodynamic Equilibrium

$$ \frac{dS}{dt} = 0 $$

$$ v = 0 $$

## Steady-State

$$ \frac{dS}{dt} = 0 $$

$$ v \neq 0 $$

## Transient

$$ \frac{dS}{dt} \neq 0 $$

$$ v \neq 0 $$

# Thermodynamic Equilibrium

Closed system (no mass transfer in or out)

    S1 -> S2;
    S2 -> S1;

````python
# Closed system
cell = '''
    S1 -> S2; k1 * S1 - k2 * S2;
    
    S1 = 10; S2 = 0;
    k1 = 1; k2 = 0.5;
'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 6, 100)
p = te.plotWithLegend(rr, result)
````

# Steady-State

Open system (mass transfer in/out)

```
$X0 -> S1;
S1 -> S2;
S2 -> $X1;
```
    
`$`: boundary species

````python
cell = '''
    $X0 -> S1; v0;
    S1 -> S2; k1*S1 - k2*S2;
    S2 -> $X1; k3*S2;

    v0 = 1;
    k1 = 2; k2 = 0; k3 = 3;
'''

rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 6, 100)
p = te.plotWithLegend(rr, result)
````

# System Response to Perturbations

Open System

$\uparrow$S1: Positive Perturbation

$\downarrow$S1: Negative Perturbation

````python
cell = '''
model cell
    $X0 -> S1; v0;
    S1 -> S2; k1*S1;
    S2 -> $X1; k2*S2;
    
    S1= 0.5;
    S2 = 0.333;
    v0 = 1;
    k1 = 2; k2 = 3;
    
    at (time > 40): S1 = 0.6; # Positive perturbation
    at (time > 20): S1 = 0.4; # Negative perturbation
end
￼￼￼￼￼￼￼￼'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 60, 300)
p = te.plotWithLegend(rr, result)
````

Closed System (no mass transfer in/out)

````python
cell = '''
model cell
    S1 -> S2; k1*S1 - k2*S2;
    
    S1 = 5;
    k1 = 1; k2 = 2;
    
    at (time > 20): S1 = 4; # Positive perturbation
    at (time > 40): S1 = 2; # Negative perturbation
end
￼￼￼￼￼￼￼￼'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 60, 300)
p = te.plotWithLegend(rr, result)
````

Note: Equilibrium values change from perturbations when there is a conserved cycle
