# System Kinetics

## Thermodynamic Equilibrium

$$ \frac{dS}{dt} = 0 $$

$$ v = 0 $$

## Steady-State

$$ \frac{dS}{dt} = 0 $$

$$ v \neq 0 $$

## Transient

$$ \frac{dS}{dt} \neq 0 $$

$$ v \neq 0 $$

# Equilibrium

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
import tellurium as te
# Simulation of a simple closed system
cell = '''
model pathway
    A -> B; k1 * A;
    B -> A; k2 * B;
    A = 10; B = 0;
    k1 = 1; k2 = 0.5;
end
'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 6, 100)
p = te.plotWithLegend(rr, result)
```

# Steady-State

```python
# Simulation of an open system
cell = '''
model pathway
    $X0 -> S1; v0;
    S1 -> S2; k1*S1 - k2*S2;
    S2 -> $X1; k3*S2;
    
    v0 = 1;
    k1 = 2; k2 = 0; k3 = 3;
end
￼￼￼￼￼￼￼￼'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 6, 100)
p = te.plotWithLegend(rr, result)
```

# System Response to Perturbations

```python
cell = '''
model pathway
    $X0 -> S1; v0;
    S1 -> S2; k1*S1;
    S2 -> $X1; k2*S2;
    
    S1= 0.5;
    S2 = 0.333;
    v0 = 1;
    k1 = 2; k2 = 3;
    
    at (time > 40): S1 = 0.6; # Positive perturbation
    at (time > 20): S1 = 0.4; # Negative perturbation
end
￼￼￼￼￼￼￼￼'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 60, 300)
p = te.plotWithLegend(rr, result)
```

## Affecting Steady State with Perturbations

```python
cell = '''
model pathway
    S1 -> S2; k1*S1 - k2*S2; # Conserved cycle
    
    S1 = 5;
    k1 = 1; k2 = 2;
    
    at (time > 20): S1 = 4; # Positive perturbation
    at (time > 40): S1 = 2; # Negative perturbation
end
￼￼￼￼￼￼￼￼'''
rr = te.loadAntimonyModel(cell)
result = rr.simulate(0, 60, 300)
p = te.plotWithLegend(rr, result)
```
